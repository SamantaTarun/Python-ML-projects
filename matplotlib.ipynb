```python
import numpy as np
from matplotlib import pyplot as plt

#Line plot
```

```python
x=np.arange(1,11)
```

```python
y=3*x
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('X vs. Y')
plt.xlabel('This is x axis')
plt.ylabel("This is y axis")
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

```python
#bar plot
students={"Bob":87,"Matt":56,"Sam":27}
names=list(students.keys())
values=list(students.values())

plt.bar(names,values,color='violet')
plt.title("Distribution of Student marks")
plt.xlabel("Names of student")
plt.ylabel("Marks of Students")
plt.show()
```

```python
#horizontal bar chart
students={"Bob":87,"Matt":56,"Sam":27}
names=list(students.keys())
values=list(students.values())

plt.barh(names,values,color='g')
plt.title("Distribution of Student marks")
plt.xlabel("Names of student")
plt.ylabel("Marks of Students")
plt.show()
```

```python
#Scatter plot
x=[4,2,3,5,8,6,2,7,9,10]
y=[6,3,4,5,2,1,7,8,6,8]

plt.scatter(x,y,marker='*',c='g',s=100)
plt.show()
```

```python
x=[4,2,3,5,8,6,2,7,9,10]
y=[6,3,4,5,2,1,7,8,6,8]
z=[2,8,9,7,1,4,5,3,10,5]

plt.scatter(x,y,marker='*',c='g',s=100)
plt.scatter(x,z,c='r',s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='g',s=100)

plt.subplot(1,2,2)
plt.scatter(x,y,marker='*',c='r',s=100)

plt.show()
```

```python
#Box plt
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data=list([one,two,three])

plt.boxplot(data)
plt.show()
```

```python
#Pi chart
fruit=['Apple','Oranges','Mango','Pineapple']
quantity=[20,10,35,20]
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','green','blue','red'])
plt.show()
```

```python
#donor chart
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%',colors=['yellow','green','blue','red'])
plt.pie([100],radius=1,colors='white')
plt.show()
```
